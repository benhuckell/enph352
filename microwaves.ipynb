## Prelab Simulations

$ \eta $ is known as the intrinsic impedance of a medium. In this case, for free space we will denote the intrinsic impedance of free space as $ \eta_0 $.

In general, this value is dependent on the **permittivity**, **permeability**, and **conductivity** of the medium.

### Derivation

Lets consider a uniform plane wave travelling in the y-direction. This means that the electric field will vary in the z-direction, and the magnetic field in the x-direction.

Starting with our electric field differential equation (in 1D), in a lossless medium (free space).

$$ \frac{\partial^2 E_x}{\partial y^2} + k^2E_x = 0 $$

This has a solution

$$ E = E_0 e^{-jky} \hat{z}$$

This represents a wave travelling with speed

$$ v_p = \frac{dy}{dt} = \frac{\omega}{k} = \frac{1}{\sqrt{\mu \epsilon}} $$


Next, using Maxwell's equations, we can derive the magnetic field $H$ as

$$ H = \frac{k}{\omega \mu} E_0 e^{-jky} \hat{x} $$

Now, since $ \eta = \frac{E}{H} $, and $ k = \omega \sqrt{\mu \epsilon} $, we can write

$$ \eta = \frac{\omega \mu}{k} = \frac{\omega \mu}{\omega \sqrt{\mu \epsilon}} = \sqrt{\frac{\mu}{\epsilon}} $$

### Free Space Calculation

$$ \epsilon_0 = 8.85 * 10^{-12} \frac{F}{m}  \hspace{1cm} \mu_0 = 4\pi * 10 ^{-7} \frac{N}{A^2} $$

$$ \eta_0 = \sqrt{\frac{\mu_0}{\epsilon_0}} = 377 \Omega $$

### How?

Since the conductivity $ \sigma $ of free space is zero, $\eta_0$  is simply determined by the ratio of the permeability of the medium to the permittivity of the medium, which will be constant.

### Electric Field Reflection Coefficient

The reflection coefficient describes the proportion of a plane wave that is reflected by an impedance discontinuity. In our case, since $ \eta_1 \neq \eta_2 $, there will be a discontinuity.

$$ \Gamma = \frac{\eta_2 - \eta_1}{\eta_2 + \eta_1} $$

Thus, the proportion of the wave that is reflected/transmitted will be dependent on both the absolute values and difference between the intrinsic impedance of each medium/material.
